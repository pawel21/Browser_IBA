```python
from plotly.subplots import make_subplots
import pandas as pd

import plotly.express as px
import plotly.graph_objects as go
from ipywidgets import widgets
from IPython.display import display
```

```python
df = pd.read_csv("data_test.csv")
df.head()
```

```python


# Create dropdown widgets for 'nazwa_polska' and 'nazwa_ostoi'
nazwa_polska_dropdown = widgets.Dropdown(
    options=df['nazwa_polska'].unique(),
    description='Nazwa Polska:',
    disabled=False,
)

nazwa_ostoi_dropdown = widgets.Dropdown(
    options=df['nazwa_ostoi'].unique(),
    description='Nazwa Ostoi:',
    disabled=False,
)

# Function to update the plot and table based on dropdown selection
def update_plot_and_table(nazwa_polska, nazwa_ostoi):
    df_filtred = df[(df['nazwa_polska'] == nazwa_polska) & (df['nazwa_ostoi'] == nazwa_ostoi)]
    
    # Create the plot
    fig = go.Figure()
    fig.add_trace(go.Bar(x=df_filtred['rok'], y=df_filtred['liczba_par_min'], name='Liczba par min'))
    fig.add_trace(go.Bar(x=df_filtred['rok'], y=df_filtred['liczba_par_max'], name='Liczba par max'))
    fig.update_layout(title=f'Liczba par min i max w poszczególnych latach dla {nazwa_polska} w {nazwa_ostoi}', 
                      xaxis_title='Rok', 
                      yaxis_title='Liczba par', 
                      barmode='group')
    fig.show()
    
    # Display the table
    display(df_filtred)

# Display the dropdowns and set up the interaction
ui = widgets.VBox([nazwa_polska_dropdown, nazwa_ostoi_dropdown])
out = widgets.interactive_output(update_plot_and_table, {'nazwa_polska': nazwa_polska_dropdown, 'nazwa_ostoi': nazwa_ostoi_dropdown})

display(ui, out)
```
